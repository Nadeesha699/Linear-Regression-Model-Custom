```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import Ridge
from sklearn.metrics import mean_squared_error, r2_score
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

company = pd.read_csv("/content/company_.csv")

x = company['Year']
x_new = x.values.reshape(-1, 1)
y = company['No of Employee']

x_train, x_test, y_train, y_test = train_test_split(x_new, y, test_size=0.2, random_state=42)

model = LinearRegression()
model.fit(x_train, y_train)

y_pred = model.predict(x_test)

print("\033[1;33mPredicted:\033[0m", "\033[1;34m", y_pred, "\033[0m", "\n")


r2 = r2_score(y_test, y_pred)
mse = mean_squared_error(y_test, y_pred)

print("\033[92mR-squared:\033[0m", r2, "\n")
print("\033[91mMean Squared Error:\033[0m", mse, "\n")


font_properties1 = {'family': 'serif', 'style': 'italic', 'size': 14}
font_properties2 = {'family': 'serif',  'size': 12}
fig = plt.figure()
fig.patch.set_facecolor('xkcd:black')
plt.scatter(x_test, y_test, color='blue', label='Actual')
plt.plot(x_test, y_pred, color='red', linewidth=2, label='Predicted')
plt.xlabel('Year', color='white',fontproperties=font_properties2)
plt.ylabel('No of Employee', color='white',fontproperties=font_properties2)
plt.title('Number of employees in a company changes over time, taking into account different years\n\n\n',color='white',fontproperties=font_properties1)
plt.tick_params(axis='x', colors='gold')
plt.tick_params(axis='y', colors='gold')
plt.grid(True, linestyle='--', alpha=0.7)
plt.legend(loc='upper left', fontsize=12)
ax = plt.gca()
ax.set_facecolor('#f0f0f0')
plt.tight_layout()
plt.legend()
plt.show()


```
